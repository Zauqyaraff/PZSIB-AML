Import pandas and numpy package

```python
import pandas as pd
import numpy as np
```

Create DataFrame by using Dictionary Methods. Using "np.nan" for null values.

```python
df1 = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"],
    "weight": [45.2,50.5,62.1,64.0,59.0,np.nan],
    "height": [155,160,162,170,167,159],
    "birth_year":[2000,1997,1996,1987,1998,2003],
    "gender":["female", "male", "male", "male", "male", "female"],
    "city":["New York", "Chicago", "Austin", "New York",np.nan, "New York"],
    })
df1
```

Create the second DataFrame consists of nickname and fullname.

```python
df2 = pd.DataFrame({
    "nick_name": ["Anna", "David", "Dane", "Rose"],
    "full_name": ["Anna Levis", "David John", "Dane Sebastian", "Rose Kimberly"],
    })
df2
```

Merge the two previous DataFrames without erasing any row and match the name and nick_name between two DataFrames by using left join method.

```python
df3 = df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name")
df3
```

Replace any null values in "full_name" column with the corresponding value of "name" column. Erase the 'nick_name' column.

```python
df3.full_name.fillna(df3.name, inplace=True)
del df3['nick_name']
df3
```

Create a new ‘BMI’ column calculated based on the following formula using values from "weight" and "height" columns.

```python
df3['BMI'] = df3['weight']/((df3['height']/100)**2)
df3
```

Create classification from "BMI" column according to given category using apply() method. Each classification indicated in the new column named "BMI_class".

```python
def categorise(row):  
    if row['BMI'] < 18.5:
        return 'Underweight'
    elif row['BMI'] >= 18.5 and row['BMI'] < 25:
        return 'Normal weight'
    elif row['BMI'] >= 25  and row['BMI'] < 30:
        return 'Overweight'
    elif row['BMI'] >= 30  and row['BMI'] < 35:
        return 'Obesity class 1'
    elif row['BMI'] >= 35  and row['BMI'] < 40:
        return 'Obesity class 2'
    elif row['BMI'] >= 40:
        return 'Obesity class 3'
    return np.nan
df3['BMI_class'] = df3.apply(lambda row: categorise(row), axis=1)
df3
```
