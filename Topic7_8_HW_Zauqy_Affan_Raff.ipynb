##### Nomor 3a:

```python
# visualisasi scatterplot menggunakan Sub Plots
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(8,10))
ax1.set_title("Penguin's Body Mass and Flipper Length Comparison")
ax1.set_xlabel("Body Mass (g)")
ax1.set_ylabel("Flipper Length (mm)")
sns.scatterplot(x = df3['body_mass_g'], y = df3['flipper_length_mm'], hue = df3['island'], ax=ax1)


ax2.set_title("Penguin's Body Mass and Bill Length Comparison")
ax2.set_xlabel("Body Mass (g)")
ax2.set_ylabel("Bill Length (mm)")
sns.scatterplot(x = df3['body_mass_g'], y = df3['bill_length_mm'], hue = df3['island'], ax=ax2)

plt.tight_layout()
plt.show()
```
